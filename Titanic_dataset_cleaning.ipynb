```python
sns.boxplot(data['Age'])
```

```python
# here we show there are many outliers are available in our age column we need to handle it or clean it
```

```python
q1=data['Age'].quantile(0.25)
q3=data['Age'].quantile(0.75)
iqr=q3-q1
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr

data['Age']=data['Age'].clip(lower=lower_limit,upper=upper_limit)
```

```python
sns.boxplot(data['Fare'])
```

```python
# here we show there are many outliers are available in our age column we need to handle it or clean it
q1=data['Fare'].quantile(0.25)
q3=data['Fare'].quantile(0.75)
iqr=q3-q1
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr

data['Fare']=data['Fare'].clip(lower=lower_limit,upper=upper_limit)
```
